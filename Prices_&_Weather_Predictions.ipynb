```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()

core_weather["month_day_max"] = core_weather["month_max"] / core_weather["temp_max"]

core_weather["max_min"] = core_weather["temp_max"] / core_weather["temp_min"]
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"12/31/2023"]
    test = core_weather.loc["1/1/2024":]

    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])

    error = mean_squared_error(test["target"], predictions)

    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```

```python
predictors = ["precip", "temp_max", "temp_min", "month_day_max", "max_min"]

error, combined = create_predictions(predictors, core_weather, reg)
error
```

# Fetch monthly weather readings from https://dev.meteostat.net/
# This website provides historical weather and climate data

```python
# Install the meteostat package
!pip install meteostat
from meteostat import Stations
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from meteostat import Monthly
```

```python
# Define latitude and longitude for Miami
latitude = 25.7617
longitude = -80.1918
```

```python
# Get nearby weather stations in Florida
stations = Stations()
stations = stations.nearby(latitude, longitude)
station = stations.fetch(1)  # Fetch data for the closest station
```

```python
# Print DataFrame
station
```

```python
# Set the station ID and time period
station_id = '72202' # Station ID for weather station near/in Miami
start = datetime(1980, 1, 1)
end = datetime(2023, 12, 30)
```

```python
# Get Monthly data for the specified station
data = Monthly(station_id, start, end)
data = data.fetch()
```

```python
# Handle missing values
data = data.fillna(method='ffill').fillna(method='bfill')
```

```python
# Temperature is measured in °Celsius
# Temperature variability (difference between max and min)
data['temp_variability'] = data['tmax'] - data['tmin']
```

```python
# Indicator for extreme temperatures (can be adjusted as per local climate)
data['extreme_cold'] = data['tavg'] < 10  # True if avg temperature is below 10°C
data['extreme_heat'] = data['tavg'] > 30  # True if avg temperature is above 30°C
```

```python
# Month and Year from the index
data['month'] = data.index.month
data['year'] = data.index.year
data['season'] = data.index.month % 12 // 3 + 1  # 1: Winter, 2: Spring, 3: Summer, 4: Fall

# Determine appropriate bin edges for temperature categories
# Set fixed bin edges that cover the expected range of temperatures
bins = [-10, 10, 20, 30, 40, 50]
labels = ['Very Cold', 'Cold', 'Mild', 'Warm', 'Hot']  # 5 labels for 5 bins
data['temp_category'] = pd.cut(data['tavg'], bins=bins, labels=labels, include_lowest=True)
```

```python
data['rolling_avg_temp'] = data['tavg'].rolling(window=3).mean()  # 3-month rolling average
data['lag_temp'] = data['tavg'].shift(1)  # previous month's temperature
data['temp_change'] = data['tavg'].diff()
```

```python
# Drop unnecessary column
data.drop(columns=['tsun'], inplace=True)
```

```python
# Sort data in descending order by time
data.sort_index(ascending=False, inplace=True)
data.sort_index(axis=1)

# Calculate mean only for numeric columns
numeric_mean = data.mean(numeric_only=True)
```

```python
# Fill NaN values with the calculated mean
data.fillna(numeric_mean, inplace=True)
```

```python
# Print DataFrame
data
```

```python
# Pivot the DataFrame for the heatmap
pivot_df = data.pivot_table(index='year', columns='month', values='tavg', aggfunc='mean')
plt.figure(figsize=(12, 11))
sns.heatmap(pivot_df, cmap="coolwarm", annot=True, fmt=".1f", linewidths=.5)
plt.title('Average Monthly Temperature in Miami (°C)')
plt.ylabel('Year')
plt.xlabel('Month')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(data['rolling_avg_temp'], color='blue', alpha=0.7)
plt.title('3-Month Rolling Average Temperature in Miami')
plt.xlabel('Date')
plt.ylabel('Average Temperature (°C)')
plt.show()
```

```python
monthly_avg_prcp = data.groupby('month')['prcp'].mean()

plt.figure(figsize=(10, 6))
monthly_avg_prcp.plot(kind='bar', color='skyblue')
plt.title('Average Monthly Precipitation in Miami')
plt.xlabel('Month')
plt.ylabel('Precipitation (mm)')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()

monthly_avg_tavg = data.groupby('month')['tavg'].mean()

plt.figure(figsize=(10, 6))
monthly_avg_tavg.plot(kind='bar', color='skyblue')
plt.title('Average Monthly Temperature in Miami')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```
